```python
import numpy as np
import pandas as pd

from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

from sklearn.model_selection import GridSearchCV
```

```python
data = pd.read_csv('./processed_data.csv')
```

```python
x =  data.drop('SalePrice', axis=1)
y = data['SalePrice']
```

```python
gbm = GradientBoostingRegressor()
param_grid = {
    'n_estimators': [30, 40, 50],
    'learning_rate': [0.1,0.2],
    'max_depth': [5, 6, 7,8],
}
x_train, x_test, y_train, y_test = train_test_split(x,y)
clf = GridSearchCV(gbm, param_grid = param_grid,scoring = 'r2')
clf.fit(x_train,  y_train)
```

```python
clf.score(x_test,y_test)
```

```python
gbm.fit(np.array(x_train[['OverallQual','TotalSF']]),y_train)
gbm.score(np.array(x_test[['OverallQual','TotalSF']]),y_test)
```

```python
from sklearn.inspection import permutation_importance
import matplotlib.pyplot as plt
```

```python
feature_importance = clf.best_estimator_.feature_importances_
sorted_idx = np.argsort(feature_importance)
pos = np.arange(sorted_idx.shape[0]) + .5
fig = plt.figure(figsize=(30, 30))
plt.barh(pos, feature_importance[sorted_idx], align='center')
plt.yticks(pos, np.array(x.columns)[sorted_idx])
plt.title('Feature Importance (MDI)')
result = permutation_importance(clf.best_estimator_, x_test, y_test, n_repeats=10,
                                random_state=42, n_jobs=2)
sorted_idx = result.importances_mean.argsort()
fig.tight_layout()
plt.show()
```
